```python
import os
import sagemaker
from sagemaker import get_execution_role
from sagemaker.tensorflow import TensorFlow

## Training

# Resources
INST_COUNT=1
INST_TYPE='ml.m4.xlarge'

sess = sagemaker.Session()
role = get_execution_role()
region = sess.boto_session.region_name
s3_train_uri = 's3://sagemaker-amzn-ti-dev/jkiggins_side_rotate/'
source_dir = os.path.join(os.getcwd(), "src_final_release")

print("source_dir: ", source_dir)

# Init and Train
mnv2_est = TensorFlow(
    entry_point='train.py',
    source_dir=source_dir,
    role=role,
    train_instance_count=INST_COUNT,
    train_instance_type=INST_TYPE,
    framework_version='1.15.0',
    py_version='py3')

mnv2_est.fit(s3_train_uri)


print("Compiling With NEO")

## Neo Compilation
input_shape = {'input_1': [4,224,224,3]} # batch size, 224x224, rgb -> 3 channels

output_path = '/'.join(mnv2_est.output_path.split('/')[:-1])
mnv2_est_optim = mnv2_est.compile_model(
     target_instance_family='sitara_am57x', 
#    target_instance_family='ml_m4',
    input_shape=input_shape,
    output_path=output_path,
    framework='tensorflow', framework_version='1.15.0')
```
